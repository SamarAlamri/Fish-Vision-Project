```python
import os
import shutil
from sklearn.model_selection import KFold
from ultralytics import YOLO
import yaml
from glob import glob

# CONFIG
DATASET_DIR = r"C:\Users\96650\Downloads\Aquarium Object Detection Dataset"
IMAGES_DIR = r"C:\Users\96650\Downloads\Aquarium Object Detection Dataset\train\images"
LABELS_DIR = r"C:\Users\96650\Downloads\Aquarium Object Detection Dataset\train\labels"
IMAGES_DIR2 = r"C:\Users\96650\Downloads\Aquarium Object Detection Dataset\test\images"
LABELS_DIR2 = r"C:\Users\96650\Downloads\Aquarium Object Detection Dataset\test\labels"
IMAGES_DIR3 = r"C:\Users\96650\Downloads\Aquarium Object Detection Dataset\valid\images"
LABELS_DIR3 = r"C:\Users\96650\Downloads\Aquarium Object Detection Dataset\valid\labels"

K = 2  # Number of folds
MODEL_NAME = r"C:\Users\96650\Desktop\ÿ∞ŸÉÿßÿßÿßÿ°2\assig2 AI2\runs\detect\train8\weights\best.pt"

# Load YAML file to get class names
with open(f"{DATASET_DIR}/data.yaml", "r") as f:
    data_yaml = yaml.safe_load(f)

class_names = data_yaml["names"]
num_classes = len(class_names)
print("Classes:", class_names)


# Collect all images
all_images = glob(f"{IMAGES_DIR}/*.jpg") + glob(f"{IMAGES_DIR}/*.png") + glob(f"{IMAGES_DIR2}/*.jpg") + glob(f"{IMAGES_DIR2}/*.png") + glob(f"{IMAGES_DIR3}/*.jpg") + glob(f"{IMAGES_DIR3}/*.png")
print(f"Total images found: {len(all_images)}")


# Fold Split
kf = KFold(n_splits=K, shuffle=True, random_state=42)

fold_results = []

for fold, (train_idx, val_idx) in enumerate(kf.split(all_images)):
    print(f"\n==============================")
    print(f"üîµ Fold {fold+1}/{K}")
    print(f"==============================")

    # --- Create fold directories ---
    fold_path = f"fold_{fold+1}"
    if os.path.exists(fold_path):
        shutil.rmtree(fold_path)

    os.makedirs(f"{fold_path}/images/train", exist_ok=True)
    os.makedirs(f"{fold_path}/images/val", exist_ok=True)
    os.makedirs(f"{fold_path}/labels/train", exist_ok=True)
    os.makedirs(f"{fold_path}/labels/val", exist_ok=True)

    # --- Split images ---
    for idx in train_idx:
        img_path = all_images[idx]
        label_path = img_path.replace("images", "labels").replace(".jpg", ".txt").replace(".png", ".txt")

        shutil.copy(img_path, f"{fold_path}/images/train/")
        shutil.copy(label_path, f"{fold_path}/labels/train/")

    for idx in val_idx:
        img_path = all_images[idx]
        label_path = img_path.replace("images", "labels").replace(".jpg", ".txt").replace(".png", ".txt")

        shutil.copy(img_path, f"{fold_path}/images/val/")
        shutil.copy(label_path, f"{fold_path}/labels/val/")

    # --- Create fold YAML file ---
    yaml_path = f"{fold_path}/data.yaml"
    fold_yaml = {
        "train": f"{fold_path}/images/train",
        "val": f"{fold_path}/images/val",
        "names": class_names,
        "nc": num_classes
    }

    with open(yaml_path, "w") as f:
        yaml.dump(fold_yaml, f)
    
    #there was error that i have to change the path inside each data.yaml manually for the code to work

    
```

```python
# fold 2
model = YOLO("yolo11n.pt")  

results = model.train(
    data = yaml_path,
    epochs = 10,
    imgsz = 640,
    batch = 16,
    project=f"cv_results",
    name=f"fold_{fold+1}",
    exist_ok=True
)
```
